P: Predict the cost of a house based on the average number of rooms

```python
# Import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import fetch_california_housing
```

## Load the California Housing dataset

```python
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns = california.feature_names)
df.head()
```

```python
df["PRICE"] = california.target
```

```python
# The target variable is the median house value for California district, expression in hundreds of thousands of $
california.target
```

## Split the data into training/testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(df[["AveBedrms"]], df["PRICE"], test_size = 0.2, random_state = 0)
```

## Create and fit the model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Make predictions using the testing set

```python
y_pred = model.predict(X_test)
```

```python
# The coefficients
print("Intercept: ", model.intercept_)
print("Coefficient: ", model.coef_)
```

# Visualizing the Regression Line

```python
plt.scatter(X_test, y_test, color = "black")
plt.plot(X_test, y_pred, color = "blue", linewidth = 3)
plt.xlabel("Number of Rooms")
plt.ylabel("House Cost")
plt.title("Average Number of Rooms vs. House Cost")
plt.show()
```

# Calculate the Model Performance

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: %.2f" % mse)
print("Coefficients of determination: %.2f" % r2)
```

As can be seen clearly from the graph, the simple linear model is not a good model for the data
